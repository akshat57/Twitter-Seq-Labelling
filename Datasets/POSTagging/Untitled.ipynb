```python
def get_pos_to_words(dataset, mapping = None):

    pos_to_words = {}
    for (tweet, labels) in dataset:
        for (word, label) in zip(tweet, labels):
            word = word.lower()

            if mapping != None:
                ##Correct this
                if label in mapping:
                    label = mapping[label]
                else:
                    continue
            
            if label not in pos_to_words:
                pos_to_words[label] = [word]
            else:
                pos_to_words[label].append(word)
                
    return pos_to_words
```

```python
gimpel_pos2words = get_pos_to_words(gimpel_data, mapping = gimpel_to_tweebank)
ritter_pos2words = get_pos_to_words(ritter_data, mapping = ritter_to_tweebank)
tweebank_pos2words = get_pos_to_words(tweebank_data, mapping = None)
```
